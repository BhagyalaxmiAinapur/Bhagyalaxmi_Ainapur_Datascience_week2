# Task 1: Basic Operations

```python
# Accept two integer inputs
num1 = int(input("Enter first integer: "))
num2 = int(input("Enter second integer: "))

# Performing arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2 if num2 != 0 else "Undefined (division by zero)"
modulus = num1 % num2 if num2 != 0 else "Undefined (modulus by zero)"
exponent = num1 ** num2

# Printing results using f-strings
print("\n--- Arithmetic Operations ---")
print(f"Addition: {num1} + {num2} = {addition}")
print(f"Subtraction: {num1} - {num2} = {subtraction}")
print(f"Multiplication: {num1} * {num2} = {multiplication}")
print(f"Division: {division}")
print(f"Modulus: {modulus}")
print(f"Exponentiation: {num1} ** {num2} = {exponent}")
```

# Task 2: Working with Lists and Arrays

```python
import numpy as np

# Creating a list of 10 numbers
numbers = [12, 45, 23, 67, 89, 34, 22, 10, 90, 55]

print("\n--- List Operations ---")
print("List:", numbers)

# Length
print("Length of list:", len(numbers))

# Max and Min
print("Maximum value:", max(numbers))
print("Minimum value:", min(numbers))

# Add and Remove elements
numbers.append(100)       # add
numbers.remove(23)        # remove

print("List after adding 100 and removing 23:", numbers)

# Sorting
numbers_sorted_asc = sorted(numbers)
numbers_sorted_desc = sorted(numbers, reverse=True)

print("Ascending order:", numbers_sorted_asc)
print("Descending order:", numbers_sorted_desc)
```

# Task 3: Dictionaries and Sets

```python
print("\n--- Dictionary and Set Operations ---")

# Create dictionary
student = {
    "name": "Bhagyalaxmi",
    "age": 22,
    "course": "Data Science",
    "marks": 85
}

# Print key-value pairs
for key, value in student.items():
    print(f"{key}: {value}")

# Add new key
student["grade"] = "A"
print("\nUpdated Dictionary:", student)

# Create a set of unique courses
courses = {"Python", "Data Science", "AI", "Python"}  # duplicates removed
print("Unique Courses Set:", courses)

# Set operations
setA = {1, 2, 3, 4}
setB = {3, 4, 5, 6}

print("Union:", setA | setB)
print("Intersection:", setA & setB)
print("Difference (A - B):", setA - setB)
```

# Task 4: File Handling

```python
print("\n--- File Handling ---")

# Create and write to file
with open("student_data.txt", "w") as f:
    f.write("Asha, Data Science, 76\n")
    f.write("Ravi, Python, 65\n")
    f.write("Kiran, AI, 89\n")
    f.write("Meena, ML, 92\n")
    f.write("Sanjay, Deep Learning, 55\n")

# Read file and display students with marks > 75
print("Students with marks above 75:")

with open("student_data.txt", "r") as f:
    for line in f:
        name, course, marks = line.strip().split(", ")
        if int(marks) > 75:
            print(name, "-", course, "-", marks)
```
